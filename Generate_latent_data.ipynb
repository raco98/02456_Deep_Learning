```python
import numpy as np
from matplotlib import pyplot as plt
import torch
import torch.nn as nn
from VAE_128 import ConvolutionalVariationalAutoencoder
```

```python
trained_network_path = "model_26_e3_0.75_200epochs\model_weights_26_e3_0.75_200epochs.pth"

latent_features = 26
cvae = ConvolutionalVariationalAutoencoder(128, latent_features)
cvae.load_state_dict(torch.load(trained_network_path,map_location=torch.device('cpu')))
```

```python
images = np.load('data_numpy\density_large128.npy')
x = torch.tensor(images, dtype=torch.float32)

if x.ndim == 3:
    x = x.unsqueeze(1)

with torch.no_grad():
    qz = cvae.posterior(x)
    latent_features = qz.mu.cpu().numpy()


np.save('latent_features501.npy',latent_features)
```
